```python
iqr= df['Sale Price'].quantile(0.75)-df['Sale Price'].quantile(0.25)
```

```python
upper_limit = df['Sale Price'].quantile(0.75) +1.5*(iqr)
lower_limit = df['Sale Price'].quantile(0.25) -1.5*(iqr)

print(upper_limit)
print(lower_limit)
```

```python
def limit_imputer(value):
    if value> upper_limit:
        return upper_limit
    if value< lower_limit:
        return lower_limit
    else:
        return value
```

```python
df['Sale Price']= df['Sale Price'].apply(limit_imputer)
```

The Variable 'Ever Renovated':
 Assign 'Yes' wherever the variable Renovated Year has a non-zero value and 'No' whenever the variable
Renovated Year is 0.
 A categorical variable.

```python
import numpy as np
df1['Ever Renovate'] = np.where(df1['Renovated Year']==0, 'NO', 'Yes')
```

Years Since Renovation = Year of Sale - Year of Renovation
 For houses that were never renovated, 'Years Since Renovation' can be assigned value of 0.
 'Years Since Renovation' is a numerical variable.

```python
df1['Purchase year'] = pd.DatetimeIndex(df1['Date House was Sold']).year
```

```python
df1['Year Since Renovation'] = np.where(df1['Ever Renovate']== 'Yes', abs(df1['Purchase year']-df1['Renovated Year']), '0')
```
